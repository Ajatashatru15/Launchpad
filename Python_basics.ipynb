```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('tbt_resampled_1.csv')
plt.figure(figsize=(10, 6))
sns.histplot(data['bid_size[0]'], kde=True, color='blue', label='Best Bid Size')
sns.histplot(data['ask_size[0]'], kde=True, color='red', label='Best Ask Size')
plt.xlabel('Size')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

# Assuming you have 'bid[0]' and 'ask[0]' columns in your DataFrame
plt.figure(figsize=(10, 6))
plot_acf(data['bid[0]'], lags=50)
plt.title('Bid Price Auto-correlation')
plt.show()

plt.figure(figsize=(10, 6))
plot_acf(data['ask[0]'], lags=50)
plt.title('Ask Price Auto-correlation')
plt.show()
```
